While we can perform a Complete Case Analysis to handle the missing data for Vancouver's PRCP and DEWP as the missing data proportion is very low (i.e., < 3\%), it seems to be sensible to drop the column "GUST" as it contains way too many missing data.




For SNDP (snow depth), we will need to examine more as we are skeptical that the missingness may indicate 0 in most of the time.

```python
len(toronto_after_2020[toronto_after_2020["SNDP"] == 0])
```

We can confirm that there is no record for 0 SNDP. For this analysis, we will make the following assumption.

If the SNDP value was 999.9 for any of the previous day or next day, we will convert it to 0 (as there is a high chance that it would just indicate 0 snow amount).

If the SNDP values for the previous day and next day are present, we will convert it to the average of those two values (as it is sensible to think that snow wouldn't just simply disappear).

With this in mind, we will modify the SNDP columns.


```python
def impute_SNDP(df):
  indices = np.where(df["SNDP"] == 999.9)[0].tolist()
  for index in indices:
    if index != 0 and index != len(df) - 1:
      if df.iloc[index - 1]["SNDP"] != 999.9 and df.iloc[index + 1]["SNDP"] != 999.9:
        df.loc[index, "SNDP"] = np.mean([df.iloc[index - 1]["SNDP"], df.iloc[index + 1]["SNDP"]])
      else:
        df.loc[index, "SNDP"] = 0
    else:
      if index == 0:
        if df.iloc[index + 1]["SNDP"] != 999.9:
          df.loc[index, "SNDP"] = df.iloc[index + 1]["SNDP"]
        else:
          df.loc[index, "SNDP"] = 0
      else:
        if index == len(df) - 1:
          if df.iloc[index - 1]["SNDP"] != 0:
            df.loc[index, "SNDP"] = df.iloc[index - 1]["SNDP"]
          else:
            df.loc[index, "SNDP"] = 0
```

We will assume the following priority orders:

Tornado > Thunder > Hail > Snow > Rain > Fog > Sunny

That is, if a particular day we had rain, snow and hail, we will consider this weather as hail.

We will denote Tornado as 6, Thunder as 5, Hail as 4, Snow as 3, Rain as 2, and Fog as 1 and finally Sunny as 0.

```python
def weather_conversion(df):
  # lazy way of computing
  tornado_indices = df.index[df["FRSHTT"] % 10 != 0].tolist()
  thunder_indices = df.index[df["FRSHTT"] % 100 != 0].tolist()
  hail_indices = df.index[df["FRSHTT"] % 1000 != 0].tolist()
  snow_indices = df.index[df["FRSHTT"] % 10000 != 0].tolist()
  rain_indices = df.index[df["FRSHTT"] % 100000 != 0].tolist()
  fog_indices = df.index[df["FRSHTT"] % 1000000 != 0].tolist()
  df.loc[:, "FRSHTT"] = 0
  df.loc[fog_indices, "FRSHTT"] = 1
  df.loc[rain_indices, "FRSHTT"] = 2
  df.loc[snow_indices, "FRSHTT"] = 3
  df.loc[hail_indices, "FRSHTT"] = 4
  df.loc[thunder_indices, "FRSHTT"] = 5
  df.loc[tornado_indices, "FRSHTT"] = 6
```

```python
toronto_test = toronto_final[toronto_final["DATE"] > "2023-03-20"]
toronto_test = toronto_test.reset_index(drop=True)
```

```python
# let's test pymc model on our data
# simple derivation of multinomial distribution of weather condition using temp,
# dewp, SNDP, and PRCP from toronto data
n = toronto_test.TEMP.values.shape[0] # number of data
X_TEMP = toronto_test.TEMP.values
X_SNDP = toronto_test.SNDP.values
X_PRCP = toronto_test.PRCP.values
X_DEWP = toronto_test.DEWP.values
X_SLP = toronto_test.SLP.values
X_VISIB = toronto_test.VISIB.values
X_WDSP = toronto_test.WDSP.values
k = 7 # number of classes

y = toronto_test["FRSHTT"].values.reshape(n, 1)

with pm.Model() as WeatherModel:
# let's test pymc model on our data
n = toronto_test.TEMP.values.shape[0] # number of data
X_TEMP = toronto_test.TEMP.values
X_SNDP = toronto_test.SNDP.values
X_PRCP = toronto_test.PRCP.values
X_DEWP = toronto_test.DEWP.values
X_SLP = toronto_test.SLP.values
X_WDSP = toronto_test.WDSP.values
k = 7 # number of classes

y = toronto_test["FRSHTT"].values.reshape(n, 1)

with pm.Model(coords={"cluster": range(2)}) as WeatherModel:
    # Posterior of Snow Depth
    sigma_sndp = pm.Normal('sigma_sndp', 3, 2)
    x_sndp = pm.HalfNormal('x_sndp', sigma=sigma_sndp, observed=X_SNDP)

    # Posterior of Sea Level Pressure
    mu_slp = pm.Normal("mu_slp", 1015, 10)
    sigma_slp = pm.HalfNormal('sigma_slp', 5)
    x_slp = pm.Normal('x_slp', mu_slp, sigma_slp, observed=X_SLP)

    # Posterior of Wind Speed
    alpha_wdsp = pm.HalfNormal('alpha_wdsp', 3)
    beta_wdsp = pm.HalfNormal('beta_wdsp', 3)
    x_wdsp = pm.Gamma('x_wdsp', alpha=alpha_wdsp, beta=beta_wdsp, observed=X_WDSP)

    # Priors of humidity regression coefficients
    beta_temp = pm.Normal('beta_temp', 0, 3, shape=(n, 2))
    beta_dewp = pm.Normal('beta_dewp', 0, 3, shape=(n, 2))
    beta_prcp = pm.Normal('beta_prcp', 0, 3, shape=(n, 2))

    # mean regression of latent variable
    mu_hum = pm.math.dot(X_TEMP, beta_temp) + pm.math.dot(X_DEWP, beta_dewp) + pm.math.dot(X_PRCP, beta_prcp)
    sigma_hum = pm.HalfNormal('sigma_hum', 10, dims="cluster")
    v = pm.Dirichlet('v', a=np.ones(2), dims="cluster")

    # Distribution of latent variable
    x_hum = pm.NormalMixture('x_hum', w=v, mu=mu_hum, sigma=sigma_hum)

    # Priors for final regression
    beta_sndp = pm.Normal('beta_sndp', 0, 3)
    beta_slp = pm.Normal('beta_slp', 0, 3)
    beta_wdsp_final = pm.Normal('beta_wdsp_final', 0, 3)
    beta_hum = pm.Normal('beta_hum', 0, 3)
    y = pm.Categorical('y', p=pm.math.softmax(pm.math.dot(beta_sndp, x_sndp)
    + pm.math.dot(beta_slp, x_slp)
    + pm.math.dot(beta_wdsp_final, x_wdsp)
    + pm.math.dot(beta_hum, x_hum)), observed=y)
```

```python
with WeatherModel:
  result = pm.sample()
```
