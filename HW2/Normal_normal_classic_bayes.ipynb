## Part 1 (Derivation of Normal Posterior Distribution):

We want to show

$exp(-\frac{1}{2}(\sum_{i = 1}^{n} \phi (x_{i} - \theta)^{2} + \tau(\theta - \theta_{0})^{2})) \propto exp(-\frac{1}{2}(\tau + n\phi)(\theta - \frac{1}{\tau + n\phi}(\tau\theta_{0} + \phi \sum_{i = 1}^{n}x_{i}))^{2})$

First, we multiply $\tau + n\phi$ to numerator and denominator.

We get

$exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + n\phi \tau (\theta - \theta_{0})^{2} + \tau \phi \sum_{i = 1}^{n} (x_{i} - \theta)^{2} + n\phi^{2} \sum_{i = 1}^{n} (x_{i} - \theta)^{2}}{\tau + n\phi}))$

Then, (note that we consider terms that do not involve $\theta$ a constant)



\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + n\phi \tau (\theta - \theta_{0})^{2} + \tau \phi \sum_{i = 1}^{n} (x_{i} - \theta)^{2} + \phi^{2} \sum_{i = 1}^{n} (n x_{i}^{2} - 2n\theta x_{i} + n\theta^{2})}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{(\theta - \theta_{0})n}{2} + \frac{\sum_{i = 1}^{n} (x_{i} - \theta)^{2}}{2(\theta - \theta_{0})}) + \phi^{2} \sum_{i = 1}^{n} n x_{i}^{2} - \phi^{2} \sum_{i = 1}^{n} 2n\theta x_{i} + \phi^{2} \sum_{i = 1}^{n} n\theta^{2}}{\tau + n\phi})) 
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{(\theta - \theta_{0})^{2}n + \sum_{i = 1}^{n} (x_{i} - \theta)^{2}}{2(\theta - \theta_{0})}) + \phi^{2} \sum_{i = 1}^{n} n x_{i}^{2} + \phi^{2} ((\sum_{i = 1}^{n} x_{i})^{2} - \sum_{i = 1}^{n} 2n\theta x_{i} + \sum_{i = 1}^{n} n\theta^{2}) - \phi^{2}(\sum_{i = 1}^{n}x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{(\theta - \theta_{0})^{2}n + \sum_{i = 1}^{n} (x_{i} - \theta)^{2}}{2(\theta - \theta_{0})}) + \phi^{2} \sum_{i = 1}^{n} n x_{i}^{2} - \phi^{2}(\sum_{i = 1}^{n}x_{i})^{2} + \phi^{2} ((\sum_{i = 1}^{n} x_{i})^{2} - 2n\theta \sum_{i = 1}^{n} x_{i} + n^{2}\theta^{2})}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\phi^{2} \sum_{i = 1}^{n} n x_{i}^{2} - \phi^{2}(\sum_{i = 1}^{n}x_{i})^{2}}{\tau + n\phi})) \cdot
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{(\theta - \theta_{0})^{2}n + \sum_{i = 1}^{n} (x_{i} - \theta)^{2}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

$\propto$

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{(\theta - \theta_{0})^{2}n + \sum_{i = 1}^{n} (x_{i} - \theta)^{2}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))

=

\begin{multline}

\end{multline}

= 

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{n\theta^{2} - 2 n\theta \theta_{0} + n\theta_{0}^{2} + \sum_{i = 1}^{n} x_{i}^{2} - 2\theta \sum_{i = 1}^{n} x_{i} + \sum_{i = 1}^{n} \theta^{2}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{n\theta^{2} - 2 n\theta \theta_{0} + n\theta_{0}^{2} + \sum_{i = 1}^{n} x_{i}^{2} - 2\theta \sum_{i = 1}^{n} x_{i} + n\theta^{2}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{2n\theta^{2} - 2 n\theta \theta_{0} + n\theta_{0}^{2} + \sum_{i = 1}^{n} x_{i}^{2} - 2\theta \sum_{i = 1}^{n} x_{i}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{2n\theta^{2} - 2n\theta \theta_{0} + \sum_{i = 1}^{n} (x_{i}^{2} + \theta_{0}^{2}) - 2\theta \sum_{i = 1}^{n} x_{i}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{2n\theta^{2} - 2n\theta \theta_{0} + \sum_{i = 1}^{n} (x_{i} - \theta_{0})^{2} + 2\theta_{0}\sum_{i = 1}^{n} x_{i} - 2\theta \sum_{i = 1}^{n} x_{i}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(\frac{2(\theta - \theta_{0})(n\theta - \sum_{i = 1}^{n} x_{i}) + \sum_{i = 1}^{n}(x_{i} - \theta_{0})^{2}}{2(\theta - \theta_{0})}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}\frac{\tau\phi(\sum_{i = 1}^{n} (x_{i} - \theta_{0})^{2})}{\tau + n\phi}) \cdot

\end{multline}

$\propto$

\begin{multline}
exp(-\frac{1}{2}(\frac{\tau^{2}(\theta - \theta_{0})^{2} + 2\tau\phi(\theta - \theta_{0})(n\theta - \sum_{i = 1}^{n} x_{i}) + \phi^{2} (n\theta - \sum_{i = 1}^{n} x_{i})^{2}}{\tau + n\phi}))
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\frac{(\tau(\theta - \theta_{0}) + \phi (n\theta - \sum_{i = 1}^{n} x_{i}))^{2}}{\tau + n\phi})
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\tau + n\phi)(\frac{(\tau(\theta - \theta_{0}) + \phi (n\theta - \sum_{i = 1}^{n} x_{i}))^{2}}{(\tau + n\phi)^{2}})
\end{multline}

=

\begin{multline}
exp(-\frac{1}{2}(\tau + n\phi)(\frac{1}{\tau + n\phi}(\tau(\theta - \theta_{0}) + \phi (n\theta - \sum_{i = 1}^{n} x_{i})))^{2})
\end{multline}

=

\begin{multline}

\end{multline}

## Part 2 (Classic Bayes)

Make "Classic Bayes" Figure

1. First, we plot individual likelihood

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# simulate 3 samples from normal distribution

x_one = stats.norm(loc=0, scale=1).rvs(1)
x_two = stats.norm(loc=0, scale=1).rvs(1)
x_three = stats.norm(loc=0, scale=1).rvs(1)

# store them in an array and set up the likelihood calculation

data = np.array([x_one, x_two, x_three])
likelihood = 1
likelihood_cont = []

# define constants and hyperparameters

n = 3
theta_0, tau = 4, 1/2

# define the domain for plotting

parameter_domain = np.linspace(-10,10,1001)

# generate the individual likelihood with an appropriate shape and store them in the container

for x in data:
    likelihood_cont.append(stats.norm.pdf((x[:,np.newaxis])*np.ones(parameter_domain.shape), 
                            loc=parameter_domain, scale=1).prod(axis=0))

# calculate the likelihood of all three data    
    
for likeli in likelihood_cont:    
    likelihood *= likeli

# calculate the pdf of the prior distribution
    
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).pdf(parameter_domain) 

# define the constant for scaling up the joint distribution for a better visualization (prior * likelihood) 

c = 500

# calculate the joint distribution

joint = c*likelihood*prior 

# calculate the posterior distribution (using the exact code from the lecture)

posterior = stats.norm(loc=(tau*theta_0+data.sum())/(tau+n), scale=1/(tau+n)**0.5).pdf(parameter_domain)

# plot the individual likelihood, prior, posterior and joint distributions
plt.figure(figsize=(16,5))
for i in range(len(likelihood_cont)):
    plt.plot(parameter_domain, likelihood_cont[i], label=f"{i + 1}th likelihood function")
#plt.plot(parameter_domain, c2*likelihood, label="single likelihood (times scalar "+str(c2)+")")
plt.plot(parameter_domain, prior, label="prior")
plt.plot(parameter_domain, posterior, label="posterior")
plt.plot(parameter_domain, joint, linestyle="--", label="joint (times scalar "+str(c)+")")
plt.title("Plots for Likelihood, Joint, Prior, and Posterior distribution of Normal-Normal")
plt.legend()
plt.show()
```

2. Make the plots in log-scale

```python
# repeat the same steps as above, but convert them into log-scale

# simulate the samples from normal distribution
# set up the constants as before

x_one = stats.norm(loc=0, scale=1).rvs(1)
x_two = stats.norm(loc=0, scale=1).rvs(1)
x_three = stats.norm(loc=0, scale=1).rvs(1)

data = np.array([x_one, x_two, x_three])
likelihood = 0
likelihood_cont = []

n = 3
theta_0, tau = 4, 1/2
parameter_domain = np.linspace(-30,30,3001)

# generate individual loglikelihood

for x in data:
    likelihood_cont.append(stats.norm.logpdf((x[:,np.newaxis])*np.ones(parameter_domain.shape), 
                            loc=parameter_domain, scale=1).prod(axis=0))

# combine a single likelihood with three data    

for likeli in likelihood_cont:
    likelihood += likeli


# create the prior distribution in log-scale
    
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).logpdf(parameter_domain)

# create a joint distribution (since they are in log, we add them)

joint = likelihood + prior 




plt.figure(figsize=(16,5))
#for i in range(len(likelihood_cont)):
#    plt.plot(parameter_domain, likelihood_cont[i], label=f"{i + 1}th log-likelihood function")

# calculate posterior distribution in log-scale

posterior = stats.norm(loc=(tau*theta_0+data.sum())/(tau+n), scale=1/(tau+n)**0.5).logpdf(parameter_domain)
plt.plot(parameter_domain, likelihood, label="combined log-likelihood")
plt.plot(parameter_domain, prior, label="prior")
plt.plot(parameter_domain, posterior, label="posterior")
plt.plot(parameter_domain, joint, label = "joint")
plt.title("Plots for Likelihood, Joint, Prior, and Posterior distribution of Normal-Normal in Log Scale")
plt.legend()
```

This would mean the only difference between the posterior and the likelihood would be marginal likelihood constant which gets divided on the likelihood.
